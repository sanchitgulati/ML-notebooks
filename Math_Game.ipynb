```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the model function
def exponential_model(x, a, b, c):
    return a * np.exp(b * x) + c

# Generate sample data
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.7, 5.5, 94.9, 17.6, 31.4])

# Fit the model to the data
params, covariance = curve_fit(exponential_model, x_data, y_data)

# Extract the parameter values
a, b, c = params

# Create a function representing the fitted equation
fitted_equation = f'{a:.2f} * e^({b:.2f}x) + {c:.2f}'

# Plot the data and the fitted equation
x_fit = np.linspace(1, 5, 100)
y_fit = exponential_model(x_fit, a, b, c)

plt.scatter(x_data, y_data, label='Data')
plt.plot(x_fit, y_fit, label=f'Fitted Equation: {fitted_equation}', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the models for different types of equations

# Linear equation: y = mx + c
def linear_model(x, m, c):
    return m * x + c

# Quadratic equation: y = ax^2 + bx + c
def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

# Exponential equation: y = a * e^(bx) + c
def exponential_model(x, a, b, c):
    return a * np.exp(b * x) + c

# Sample data
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.7, 5.5, 9.9, 17.6, 31.4])

# Fit linear model
params_linear, _ = curve_fit(linear_model, x_data, y_data)

# Fit quadratic model
params_quadratic, _ = curve_fit(quadratic_model, x_data, y_data)

# Fit exponential model
params_exponential, _ = curve_fit(exponential_model, x_data, y_data)

# Extract parameter values for each model
m, c = params_linear
a, b, quadratic_c = params_quadratic
exp_a, exp_b, exp_c = params_exponential

# Create string representations of the fitted equations
fitted_linear_eq = f'{m:.2f}x + {c:.2f}'
fitted_quadratic_eq = f'{a:.2f}x^2 + {b:.2f}x + {quadratic_c:.2f}'
fitted_exponential_eq = f'{exp_a:.2f} * e^({exp_b:.2f}x) + {exp_c:.2f}'

# Generate fitted curves for plotting
x_fit = np.linspace(1, 5, 100)
y_fit_linear = linear_model(x_fit, m, c)
y_fit_quadratic = quadratic_model(x_fit, a, b, quadratic_c)
y_fit_exponential = exponential_model(x_fit, exp_a, exp_b, exp_c)

# Plot the data and the fitted equations
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_fit, y_fit_linear, label=f'Linear Equation: {fitted_linear_eq}', color='red')
plt.plot(x_fit, y_fit_quadratic, label=f'Quadratic Equation: {fitted_quadratic_eq}', color='blue')
plt.plot(x_fit, y_fit_exponential, label=f'Exponential Equation: {fitted_exponential_eq}', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
